Two common ways to measure the center of a data set are the mean and median. Both of these statistics can be used to show what a normal observation is.

But they are calculated in different ways, and so they give different information. Whether the mean and median should be used depends on context.

The mean — or average — is the sum of the values in the set divided by the number of items in the set.
The median is the value that splits a set in half when it is ordered from least to greatest.

That is, the middle number in a set of values.

```python
np.mean([2,3,4,5,10,20,40])
```

In a set with an even number of terms, the median is slightly more complicated, because there is no value in the set that is truly the middle term.

When this happens, the median is calculated by finding the mean of the middle two terms.
The mean and median are known as measures of central tendency.
Remember that extreme values will pull the mean towards them, but not the median.


```python

data = [1, 2, 3, 4, 5]
data = [0,0,1,1,1,2,2,2,2,100]
data=[1,1,10]
print("Mean is :", round(st.mean(data),5))
print("Median is :", round(st.median(data),5))
print("Mode is :", round(st.mode(data),5))
```

The most common situation in where the mean and median are identical but the mode is different is called a “bimodal” or “multimodal” distribution.It is possible for the mean to be less than the median, even in cases where the distribution is bimodal, and even in cases where the major mode is greater than the minor mode.

One way for this to happen is for there to be a large amount of data in the lower values (which has a strong effect on the mean but little effect on the median), and this will have an even greater effect if the major mode has a fairly narrow distribution.

It is worth noting that for “regular” or “normal” distributions, we would expect the median, mean, and mode to all be close to the same value. We can learn quite a bit by comparing these three values, but to further characterize the distribution of data, especially irregular data, we will need additional tools.

One such measure — quantiles — extend the idea of the median to more points in the data, which is what we'll explore in the next lesson.
In addition to minimum, maximum, and median, we can pick other different percentage points — or quantiles.


```python
data=[0, 1,1,2,3,3,3,1000]

print("Mean is :", round(st.mean(data),5))
print("Median is :", round(st.median(data),5))
print("Mode is :", round(st.mode(data),5))
st.quantiles(data)
st.quantiles(data, method='inclusive')

data=[0, 100,101, 102,103,104,105]

print("Mean is :", round(st.mean(data),5))
print("Median is :", round(st.median(data),5))
print("Mode is :", round(st.mode(data),5))
st.quantiles(data)
st.quantiles(data, method='inclusive')
```

We do not want to use the population variance formula on a sample because it is a biased estimator.

Applying the population variance formula to determine the sample variance will generally underestimate the population variance, since samples will tend to miss outliers.

However, if, in our variance formula, we divide by something smaller than the number of items in the sample, we will get a larger variance value that is closer to the population variance.

We can compensate by dividing by $n-1$, which will change our formula to get an unbiased estimate.
A particular sample still won't be perfect, but if we took every sample of the same size and averaged them we'd get the correct population variance.

In statistics, we make a distinction between observation and experiment.
An observation is simply the record of some particular value at a particular point in time.An observational study involves the collection of many observations that the investigators think are worth considering.

However, the investigators do not change anything about the situation or the variables involved in an observational study.In contrast, an experiment involves the application of specific treatments to selected groups followed by the observation of specific values.

Experiments always involve the direct (or indirect) manipulation of the situation by the researcher in an attempt to find a relationship between two different variables.

Statistical significance is a claim by researchers that the results observed in their experiment are unlikely to have been caused by random chance.

When we are performing an experiment, we expect some variation in the results between our groups just based on randomness.

Observations are simply collected data from the past or present in which there is no human intervention. An example would be survey data, where a researcher asks a list of questions. It's possible, however, that the data collection method is biased, which can negatively affect the outcome of your survey.

Experiments are manipulated scenarios in which lurking variables are controlled to draw precise conclusions. An example would be a pharmaceutical experiment, where a drug and a placebo are given to different groups that don't know which they've received.

Experiments are often more powerful because they can be controlled. But sometimes controlling experimental units may be unethical or impossible, and in this case we rely on observations.

Blocking is when experimental units are put into groups (blocks) that are similar to one another. This allows the experiment to control for lurking variables.The goal is to create an experiment where the only difference between the experimental units is the treatment of interest.

simple random sampling, where the group to study is chosen from the population at random. However, this approach is not always possible or optimal.

Blinding is one way of reducing bias in experimental studies.

The placebo effect is a good example of confounding. It is a result that happens along with our treatment that produces the same kind of effect we are hoping to measure.

size, convenience, and randomization are all important in real-world experiments.

We usually choose to block when we have reason to believe that our subjects differ in some way that influences our measurement or when we have reason to believe that our treatment will affect different groups of subjects differently.

Simple random sampling is often the easiest way to sample from a population.

Stratified sampling involves first dividing the population into groups and then choosing a random — but proportional— sample from each group.

A sample that is drawn from a small group that we believe to be representative of the population at large is called a cluster sample.

While stratified sampling usually involves more work in order to reduce the variability of our sample, cluster sampling almost always requires less work and we hope that it doesn't introduce bias into our measurements.

Systematic sampling involves choosing a set interval at which to sample from a list containing the population, rather than using a simple random sample. Systematic sampling is used primarily for convenience, not because it improves results over simple random sampling.

 In situations where there is a significant amount of variance, it sometimes makes more sense to measure the difference between a related pair of subjects instead of sampling randomly.

Type 1 - false positive - the null hypothesis is true, but we reject it- situations where we believe our theory to be true even though it is not. Probability of false positive is $\alpha$ and equal to $p$-value, which is why we choose a low $p$-value

Type 2 - false negative - the null hypothesis is false, but we fail to reject it. situations where our theory is true but our data does not support rejecting $H_O$. Probability of false negative is $\beta$. This type of error is unlikey to be published because the experiment failed to reject $H_O$, or, fail to give statistically significant evidence of effect or difference. $1-\beta$ is the power of an experiment - the probability the experiment can find the effect it is looking for assuming the effect is there. 

Rejecting $H_O$ is a positive result — it means that we are accepting the evidence as valid for our alternative hypothesis.

 a higher p-value (a lower "bar") means a higher probability that a positive results comes from random variations in the null hypothesis model.
 
Keep in mind that failing to reject the null hypothesis does not mean that the null hypothesis is true, only that there is not enough evidence to support the alternative hypothesis. Similarly, rejecting the null hypothesis does not prove that the alternative hypothesis is true, only that it is more likely given the data.
 
$p^*$ is the particular proportion we must find to reject the null hypothesis. If our experiment produces a value above $p^*$, we will reject the null hypothesis. If we find a value below 
$p^*$ , we will not reject the null hypothesis.

It seems a question's usefulness is somehow related to how much uncertainty we have about the answer. The most informative questions are the ones whose answer we are the most uncertain about.
maximally informative questions - where both answers are equally likely. A useful way to think about maximally informative questions is like flipping a fair coin. Before the coin is flipped, we have no insight into what the outcome will be — if we tried to guess, we'd be wrong half the time and right the other half of the time.
It's impossible to know less than this about the coin's outcome. This is what is meant by maximal uncertainty. If we ask what happened, we'd go from zero knowledge to complete knowledge.

information isn't a vague concept but a mathematical quantity. information can be quantified in units called bits.
